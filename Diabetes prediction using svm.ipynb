```python
#Identifying unique values so that it can be repalced.
df['Insulin'].unique()
```

```python
#Replacing the zero with mean values.
df['Insulin']=df['Insulin'].replace(0,df['Insulin'].mean())
df['Glucose']= df['Glucose'].replace(0, df['Glucose'].mean())
df['Pregnancies']= df['Pregnancies'].replace(0, df['Pregnancies'].mean())
df['SkinThickness']= df['SkinThickness'].replace(0, df['SkinThickness'].mean())
df['BMI']= df['BMI'].replace(0, df['BMI'].mean())
df['BloodPressure']= df['BloodPressure'].replace(0, df['BloodPressure'].mean())
```

```python
#Importing train and test libraries.
from sklearn.model_selection import train_test_split  
```

```python
# training the model.
x= df.drop('Outcome', axis = 1)
y= df['Outcome']
```

```python
#Splitting tha train and test data.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)  
```

```python
#Performing logistic regression model.
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(C=0.1, max_iter=30, fit_intercept=True, n_jobs=3,solver='lbfgs',multi_class='multinomial')
LR.fit(x_train, y_train)
```

```python
prediction = LR.predict(x_test)
prediction
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, prediction)
```

```python
#performing support vector machine.
from sklearn.svm import SVC  
svclassifier = SVC(C=1.0,kernel='poly',max_iter=-1,degree=2, tol=0.01, random_state=None,
                  cache_size=200,class_weight=None, probability=False, decision_function_shape='ovr', shrinking=True)  
svclassifier.fit(x_train, y_train) 
```

```python
y_pred = svclassifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix  ,accuracy_score
print(classification_report(y_test,y_pred))  
print(accuracy_score(y_test,y_pred))
```
